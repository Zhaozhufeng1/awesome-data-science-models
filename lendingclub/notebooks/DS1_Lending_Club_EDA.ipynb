```python
def expand_year(x):
    """Converts YY -> YYYY

    Assumes there are no records in the current year, which will be returned as-is
    """
    this_year = datetime.datetime.now().strftime("%y")
    if int(x[::-1][:2][::-1]) < int(this_year):
        return x[0:len(x) - 2] + "20" + x[len(x) - 2:]
    elif int(x[::-1][:2][::-1]) > int(this_year):
        return x[0:len(x) - 2] + "19" + x[len(x) - 2:]
    else:
        return x
```

```python
earliest_cr = earliest_cr.astype(str)
```

```python
text_columns = ['Notes','purpose']
text_df = data_cluster[text_columns]
```

```python
text_df['text'] = text_df.Notes +" "+ text_df.purpose
```

```python
import re
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
```

```python
regex = re.compile(r'\w+\sadded\son\s\d+\/\d+\/\d+\s>\s')
text_df['text'] = text_df['text'].apply(lambda x: regex.sub("", x))
```

```python
lemmatizer = WordNetLemmatizer()
stop = set(stopwords.words('english'))
# docs = [[lemmatizer.lemmatize(token) for token in doc] for doc in docs]

def replace_token(t):
    if t in ['consolidation','consolidating']:
        return 'consolidate'
    elif t in ['currently']:
        return 'current'
    elif t in ['financial']:
        return 'finance'
    elif t in ['paying','payment']:
        return 'pay'
    elif t in ['requesting']:
        return 'request'
    else:
        return t

def tokenizer(text):
    tokens = [word.lower() for word in nltk.word_tokenize(text) if len(word) > 1]
    tokens = [word for word in tokens if word not in stop]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = [word for word in tokens if word not in ["br/","'s","'m","--","...","u","n't"]]
    tokens = [replace_token(t) for t in tokens]
    return tokens
```

```python
tfidf = CountVectorizer(min_df=0.03,max_df=0.5, tokenizer=tokenizer)
```

```python
matrix = tfidf.fit_transform(text_df.text)
```

```python
features = tfidf.get_feature_names()
```
